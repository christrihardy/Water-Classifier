### **1. Libraries**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import src.util as utils
import pandas as pd
import numpy as np
from imblearn.over_sampling import SMOTE
```

### **2. Load Config File**

```python
config = utils.load_config()
```

### **3. Load Train set**

```python
x_train = utils.pickle_load(config["train_set_path"][0])
y_train = utils.pickle_load(config["train_set_path"][1])
```

```python
# concat input-output
train_set = pd.concat([x_train, y_train], axis = 1)
```

### **4. Data Exploration**

4.1. Quick Look Dataset

```python
# Quick take a loof to our train data
train_set
```

```python
train_set.describe()
```

4.2. Label Distribution

```python
# Check distribution of our label data
ax = sns.countplot(data = train_set, x = "Potability", label = train_set["Potability"].unique())
ax.bar_label(ax.containers[0])
```

Balance data

```python
# balance data with smote
x, y = SMOTE(sampling_strategy = "minority",
            random_state = 123).fit_resample(x_train, y_train)
train_set_bal = pd.concat([x, y], axis = 1)
```

```python
# Check balancing result
y.value_counts()
```

```python
# Check the balanced data with smote method
ax = sns.countplot(data = train_set_bal, x = "Potability", label = train_set_bal["Potability"].unique())
ax.bar_label(ax.containers[0])
```
